```
# -*- coding: utf-8 -*-

"""CS PROJECT.ipynb

Automatically generated by Colaboratory.

Original file is located at
    https://colab.research.google.com/drive/1KXv9yxMNEdcZuCqJJlLEvvnY3RNALEDy

**BLOOD BANK**

**DATABASE**
"""

import sqlite3 as s
conn=s.connect('BloodBank.db')
c=conn.cursor()

#table blood_stck
c.execute('''create table blood_stck([Sno] INTEGER PRIMARY KEY,[type] text,[stock] integer)''')
conn.commit()
#table cmpts_stck
c.execute('''create table cmpts_stck([Sno] INTEGER PRIMARY KEY,[type] text,[Qty] integer)''')
conn.commit()
#table requests
c.execute('''create table requests([Pid] INTEGER PRIMARY KEY,[Pname] TEXT,[type] TEXT,[hospital] TEXT,[units] integer,[contact
] integer,[status] text)''')
conn.commit()
#table donations
c.execute('''create table donations([Sno] int PRIMARY KEY,[Dname] text,[Dage] integer,[contact] integer,[address] text,[pin_code] integer,[status] text)''')
conn.commit()

"""**values**"""

import sqlite3 as s
conn=s.connect('BloodBank.db')
c=conn.cursor()

#input values blood_stck
c.execute('''insert into blood_stck values(1,"O+",224)''')
c.execute('''insert into blood_stck values(2,"O-",220)''')
c.execute('''insert into blood_stck values(3,"B+",200)''')
c.execute('''insert into blood_stck values(4,"B-",200)''')
c.execute('''insert into blood_stck values(5,"A+",200)''')
c.execute('''insert into blood_stck values(6,"A-",200)''')
c.execute('''insert into blood_stck values(7,"AB+",200)''')
c.execute('''insert into blood_stck values(8,"AB-",200)''')
c.execute('''insert into donations values(1,"G Shikhawat",23,8789076567,"C4 Long Island",110087,"P")''')

conn.commit()
conn.close()

import sqlite3 as s
conn=s.connect('BloodBank.db')
c=conn.cursor()

#input values cmpts_stck
c.execute('''insert into cmpts_stck values(1,"Plasma",300)''')
c.execute('''insert into cmpts_stck values(2,"RBC",250)''')
c.execute('''insert into cmpts_stck values(3,"WBC",240)''')

conn.commit()
conn.close()

"""**GUI**"""

import sqlite3 as s
conn=s.connect('BloodBank.db')
cur=conn.cursor()


def main():
  while True:
    print("**********************************************************************")
    print("\t\t\t BLOOD BANK")
    print("**********************************************************************")
    print("MAIN MENU")
    print("1. Request Blood")
    print("2. Donate")
    print("3. Administration")
    print("4. About us")
    print("5. Exit")
    ch=input("Enter your choice:  ")
    print("---------------------------------------------------------------------")
    if ch=='1':
      request()
    elif ch=='2':
      donations()
    elif ch=='3':
      admin()
    elif ch=='4':
      about_us()
    elif ch=='5':
      print("Thank You....EVERY DROP COUNTS!!!")
      break
    else:
      print("Please enter a valid choice....")



def request():
    while True:
        print("\t\tREQUEST BLOOD")
        print("Please enter all the necessary details")
        a=input("Enter Pno(4-digit):  ")
        if a.isdigit==False or len(a)!=4:
          print("Invalid Pno! Try again")
        else:
          b=input("Enter patient's name:  ")
          c=input("Enter the type of blood required(eg: O+,AB-):  ")
          if c!='A+' and c!='A-' and c!='B+' and c!='B-' and c!='AB+' and c!='AB-' and c!='O+' and c!='O-':
            print("Enter valid blood group type please")
          else:
            d=input("Enter the name of the hospital:  ")
            e=eval(input("Enter the no. of blood units required:  "))
            if e>4:
              print("Please enter valid number of blood units required[Max 4 units per patient]")
            else:
              f=input("Enter your phone number:  ")
              if len(f)!=10 or f.isdigit==False:
                print("Please enter a valid phone number")
              else:
                qry="insert into requests values({},'{}','{}','{}',{},{},'{}')".format(a,b,c,d,e,f,'P')
                cur.execute(qry)
                conn.commit()
                print("Request Lodged")
                print("Get well soon and take care of your health")
                break



def donations():
    cur.execute('select Sno from donations')
    row=cur.fetchall()
    while True:
        s=row[-1][-1]
        s=s+1
        print("WELCOME")
        print("You are one step away from saving a life")
        a=input("First of all,please confirm if you are healthy,although we'll do a blood test too(Y/N):  ") #will not be present in the table
        if a=='N':
          print("You can't donate blood now but do try again once you are healthy" )
          break
        elif a=='Y':
            print("Please enter all the necessary details")
            c=input("Enter your name: ")
            d=eval(input("Enter your age:  "))
            if d<18:
                print("Sorry! You are underage for donating blood right now.")
            else:
                e=input("Enter your phone number:  ")
                if len(e)!=10:
                    print("Please enter a valid phone number")
                else:
                    f=input('Enter your address: ')
                    g=input('Enter the pincode of the area: ')
                    if len(g)!=6 or g.isdigit==False:
                        print("Please enter a valid pincode.")
                    else:
                        qry="insert into donations values({},'{}',{},'{}','{}','{}','{}')".format(s,c,d,e,f,g,'pending')
                        cur.execute(qry)
                        conn.commit()
                        break
        else:
            print("Please enter a valid response")




def admin():
    P = input("Enter The Password: ")
    if P=="Blood":
        while True:
            print("---------------------------------------------------------------------")
            print("\t\t|||Blood Bank's Administration||| ")
            print("1. View all requests")
            print("2. View donations data")
            print("3. Modify the status(pending/completed) ")
            print("4. Update blood components stock")
            print("5. Display stocks")
            print("6. Back To Main Menu")
            ch=int(input("Enter your choice:  "))
            if ch==1:
                view_requests()
            elif ch==2:
                view_donations()
            elif ch==3:
                modify()
            elif ch==4:
                update()
            elif ch==5:
                display()
            elif ch==6:
                break
            else:
                print("Wrong Choice.....Enter Again")
    else:
        print("Wrong Password.....Try Again")
        admin()



def view_requests():
  print('These are all the placed requests:')
  cur.execute('''select * from requests''')
  rows=cur.fetchall()
  for i in rows:
    print(i)



def view_donations():
  print("These are the details of the donations:")
  cur.execute('''select * from donations''')
  dono=cur.fetchall()
  for i in dono:
    print(i)



def about_us():
   print('\n\t\tAbout Subhash Chandra Bpse Blood Bank\t\t')
   print("Subash Chandra Bose Blood Bank is a non-profitable organisation that has been saving lives since 1994\
   \n*********Aap Hume Khoon Do!Hum Aapko Zindagi Denge*************\
   \nBlood is a body fluid in humans and other animals that delivers necessary substances such as nutrients\
   \nand oxygen to the cells and transports metabolic waste products away from those same cells.Sometimes blood\
   \ntransfusion plays a crucial role in saving someone's life. We as a blood bank make the process of collection\
   \nand distribution of blood, even the rarest types, easier so that no one has to lose their life due to blood\
   \nloss or unavailability of blood.We,with the help of the generous blood donators have saved millions of lives\
   \nand hope to save millions more!")



def modify():
    print('||!!!ADMINISTRATIVE TOOL, PROCEED WITH CAUTION!!!||')
    print('Whose status would you like to change?')
    print('1. Requests')
    print('2. Donations')
    a=int(input('Enter your choice:'))
    if a==1:
      qry='select * from requests'
      cur.execute(qry)
      row=cur.fetchall()
      if len(row)==0:
        print('This table is currently empty!')
        print('Status cannot be changed!\nGoing back to menu...')
      else:
        for i in row:
          print(i)
        b=int(input("Whose status would you like to change, P.Id:"))
        c=input('Enter the new status,pending or completed? (P/C): ')
        qry1="update requests set status='{}' where Pid={}".format(c,b)
        cur.execute(qry1)
        conn.commit()
    if a==2:
      qry='select * from donations'
      cur.execute(qry)
      row=cur.fetchall()
      for i in row:
        print(i)
      b=int(input("Whose status would you like to change, S.No.:"))
      c=input('Enter the new status: (P/C) ')
      if c!='P' or c!='C':
          print("Enter a valid status")
      else:
          qry1="update donations set status='{}' where Sno={}".format(c,b)
          cur.execute(qry1)
          conn.commit()
    else:
        print("Enter a valid choice")



def update():
  print('1. Blood stocks')
  print('2. Components stocks')
  a=input('Enter your choice:')
  print("Here are all the stocks")
  if a=='1':
    cur.execute('select Sno from blood_stck')
    row=cur.fetchall()
    s=row[-1][-1]
    cur.execute('select * from blood_stck')
    row=cur.fetchall()
    for i in row:
      print(i)
    a=int(input('Whose stock would you like to update Sno:  '))
    if a<=s:
        new_stock=int(input('Enter the new value: '))
        qry="update blood_stck set stock={} where Sno={}".format(new_stock,a)
        cur.execute(qry)
        conn.commit()
    else:
        print('Enter valid Sno')
  if a=='2':
    cur.execute('select * from cmpts_stck')
    row=cur.fetchall()
    for i in row:
      print(i)
    a=int(input('Whose stock would you like to update Sno:  '))
    new_stock=int(input('Enter the new value: '))
    qry="update cmpts_stck set Qty={} where Sno={}".format(new_stock,a)
    cur.execute(qry)
    conn.commit()
  else:
    print("Enter a valid choice")



def display():
  print('1. Blood stocks')
  print('2. Components stocks')
  a=input('Enter your choice:')
  print("Here are all the stocks")
  if a=='1':
    cur.execute('''select * from blood_stck''')
    row=cur.fetchall()
    if len(row)==0:
      print('This table is currently empty!')
    else:
      for i in row:
        print(i)
  if a=='2':
    cur.execute('''select * from cmpts_stck''')
    row=cur.fetchall()
    if len(row)==0:
      print('This table is currently empty!')
    else:
      for i in row:
        print(i)
  else:
        print("Enter a valid choice")


main()
```
